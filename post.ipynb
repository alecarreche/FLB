```python
import pandas as pd
import seaborn as sns
```

```python
# load results from model
results = pd.read_csv("data/results.csv", index_col="jax_data.Coaster_Name").drop("Unnamed: 0", axis=1)

# load original data
training_data = pd.read_csv("data/normalized_coaster_data.csv", index_col="Coaster_Name")
```

```python
# get predicted top ten from the best model 
top_ten = results.sort_values(by="svm_rad", ascending=False)[:10]
original_top_ten = training_data.sort_values(by="success_metric", ascending=False)["success_metric"][:10]
```

```python
# copied from preprocessing for back transformation 
success_min = 2860.840789355851
success_max = 295477.7314716182
```

```python
 # back transform
 top_ten = top_ten * (success_max-success_min) + success_min
```

```python
# back transform 
original_top_ten = pd.DataFrame(original_top_ten * (success_max-success_min) + success_min)
```

```python
top_ten.to_csv("predicted_top_ten.csv")
original_top_ten.to_csv("original_top_ten.csv")
```
